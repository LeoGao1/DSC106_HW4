```python
from imdbpie import ImdbFacade
from IPython.core.display import display, HTML
from bs4 import BeautifulSoup
import urllib.request
import re
```

```python
# Get an instance of IMDb class
imdb = ImdbFacade()

# Search for Arnold Schwarzenegger
people = imdb.search_for_name('Leonardo DiCaprio')
print(people)
```

```python
# Fetch information about him
leo = imdb.get_name(people[0].imdb_id)

# What information do I have about him?
print('\n'.join([x for x in dir(leo) if not x.startswith('__')]))
```

```python
# Let's fetch some more information about a movie
movie = imdb.get_title(leo.filmography[-1])
```

```python
# What information can I get about this movie?
print('\n'.join([x for x in dir(movie) if not x.startswith('__')]))
```

```python
movie.year
```

```python
directors = []
genres = []
year = []
rating = []
rating_count = []
count = 0
for i in leo.filmography:
    try:
        m = imdb.get_title(i)
        for j in m.directors:
            directors.append(j.name)
        genres.append(m.genres)
        year.append(m.year)
        rating.append(m.rating)
        rating_count.append(m.rating_count)
    except:
        continue
    print(count)
    count += 1
    

print('done')
```

```python
direc_count= {}
for i in directors:
    if i in direc_count.keys():
        direc_count[i] = direc_count[i]+1
    else:
        direc_count[i] = 1

d_count = {}
for i in direc_count.keys():
    if direc_count[i] >2:
        d_count[i] = direc_count[i]
        
d_count
```

```python
genre_count= {}
for i in genres:
    for j in i:
        if j in genre_count.keys():
            genre_count[j] = genre_count[j]+1
        else:
            genre_count[j] = 1
        
g_count = []
for i in genre_count.keys():
    g_count.append([i,genre_count[i]])
    
g_count.sort(key=lambda x:x[1],reverse = True)
g_count
```

```python
event_list = []
count = 0
for i in leo.filmography:
    try:
        m = imdb.get_title(i)
        d = m.release_date
        event_list.append([m.title,d.year,d.month,d.day,m.rating])
    except:
        continue
    print(count)
    count += 1
    

print('done')
```
